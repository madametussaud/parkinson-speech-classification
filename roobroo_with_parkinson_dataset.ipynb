```python
import pandas as pd


train = pd.read_csv("/content/train_data.txt", sep=",", header=None)

train.head()
```

```python
print("Shape of dataset:", train.shape)
```

```python
X = train.iloc[:, :-1]   # all columns except last
y = train.iloc[:, -1]    # last column

print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,        # 80% train, 20% test
    stratify=y,           # keeps class balance
    random_state=42,
    shuffle=True
)
```

```python
print("Training class distribution:")
print(y_train.value_counts())

print("\nTest class distribution:")
print(y_test.value_counts())
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
import pandas as pd

# Convert numpy arrays → DataFrames
X_train_scaled_df = pd.DataFrame(X_train)
X_test_scaled_df = pd.DataFrame(X_test)

y_train_df = pd.DataFrame(y_train).reset_index(drop=True)
y_test_df = pd.DataFrame(y_test).reset_index(drop=True)
```

```python
train_scaled_full = pd.concat([X_train_scaled_df, y_train_df], axis=1)
test_scaled_full = pd.concat([X_test_scaled_df, y_test_df], axis=1)

train_scaled_full.head()
```

```python
train_scaled_full.to_csv("parkinsons_train_scaled.csv", index=False)
test_scaled_full.to_csv("parkinsons_test_scaled.csv", index=False)
```
